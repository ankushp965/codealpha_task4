```python
def detect_iqr(df):
    Q1= df['age'].quantile(0.25)
    Q3= df['age'].quantile(0.75)
    IQR= Q3-Q1
    lower_limit= Q1-1.5*IQR
    upper_limit= Q3+1.5*IQR
    return df[(df['age']<lower_limit)|(df['age']>upper_limit)]
outliers=detect_iqr(df)
print(outliers)
```
